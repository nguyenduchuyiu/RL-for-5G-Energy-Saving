```python
import numpy as np
import matplotlib.pyplot as plt

def _violation_penalty(x, th):
    violation_ratio = max(0.0, (x - th) / max(1e-6, th))
    return violation_ratio

# Parameters
th = 0.01  # 1%
x_values = np.linspace(0, 0.06, 100)  # 0 to 6%

# Calculate violation penalties
penalties = [_violation_penalty(x, th) for x in x_values]

# Create the plot
plt.figure(figsize=(10, 6))
plt.plot(x_values * 100, penalties, linewidth=2, color='red', label='Violation Penalty')
plt.axvline(x=th * 100, color='gray', linestyle='--', alpha=0.7, label=f'Threshold = {th*100}%')
plt.xlabel('x (%)')
plt.ylabel('Violation Penalty')
plt.title('Violation Penalty Function')
plt.grid(True, alpha=0.3)
plt.legend()
plt.xlim(0, 6)
plt.ylim(0, max(penalties) * 1.1)
plt.show()
```
